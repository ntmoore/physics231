```python
mass=[]
t=[]
with open("stream.csv") as csvfile:
    datafile=csv.reader(csvfile)
    for row in datafile:
        mass.append(row[0])
        t.append(row[1])
print(mass)
print(t)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(mass,t,"r+",label="Problem 2 data")
#plt.plot(mass,"g-")
plt.ylabel("time (s)")
plt.xlabel("mass (kg)")
plt.legend()
plt.show()
```

I'd like to fit by hand

```python
m_fitted=[]
t_fitted=[]
for i in range(10):
    temp_mass=i*(23/10)
    time=1.4*temp_mass+3.2
    #print(mass, "kg\t\t",time,"s")
    m_fitted.append(temp_mass)
    t_fitted.append(time)

# this is the data 
plt.plot(mass,t,"r+",label="Problem 2 data")
# here's my fit
plt.plot(m_fitted,t_fitted,label="time (s) = 1.4(s/kg)*mass(kg)+3.2s")
plt.ylabel("time (s)")
plt.xlabel("mass (kg)")
plt.legend()
plt.show()
```

```python
p=[]
v=[]
one_over_v=[]
with open("problem_9.csv") as csvfile:
    datafile=csv.reader(csvfile)
    for row in datafile:
        v.append(row[0])
        p.append(row[1])
        one_over_v.append(1.0/float(row[0]))
print(v)
print(p)
print(one_over_v)
```
